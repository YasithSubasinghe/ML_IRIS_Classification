```python
from sklearn.model_selection import train_test_split  #skelern is libarry// train_test_split devide tarin and test data
import pandas as pd  # to work with data frames
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)  # print iris attributes in dataset    //if print iris.data it contain length values according to seperate flowers
#iris.target print types of flowers values based on data values
#iris.target_name
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)  #create dataframe with using iris.features feature into the coloums
df.head()  #head(10) first 10 values
```

```python
df['target']=iris.target  #merge target columns with undepenent variables to proper understand
df.head()
```

```python
df[df.target==1].head()  #filter versicolor to display.its number is 1.That why it use  //if head() that print firat 5 element
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x]) # to identify clearly add flower name to the table. lamda use to speed execution than other technologies
df.head(100)                           #name assign using target_name Array
```

```python
df0 = df[:50] # sethosa
df1 = df[50:100] # versicolor   # data set break into three parts #df1 mean data framework
df2 = df[100:] # virginica
```

```python
import matplotlib.pyplot as plt  #matplotlib use for plottimg and it is a libary

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color="green", marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color="blue", marker='.')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color="orange", marker='.')

plt.show()
```

```python
X = df.drop(['target','flower_name'], axis='columns') #now avoid target and flower name and assign that to X.that mean feature variables
y = df.target #assign target values to y
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # divide 20% of the dataset for the testing\n",
len(X_train) #check length of X_train
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC() #assign SVC instance to model
```

```python
model.fit(X_train, y_train) #X_train=feature variable for tarining    #y_train=target variables for training
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]]) #pass sepal length petal length ..... ... ..
```

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```
